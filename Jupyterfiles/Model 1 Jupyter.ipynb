```python
# Check common columns (removed "ra" and "dec")
df1 = df1.drop(columns=["ra", "dec"], errors="ignore")
df2 = df2.drop(columns=["ra", "dec"], errors="ignore")

combined=pd.concat([df1, df2])

common_cols = set(df1.columns) & set(df2.columns)
print("Common columns after dropping ra and dec:")
print(sorted(common_cols))
```

```python
#Convert Data into a managable format for SciKit-Learn

Xvals= ['orbital_period', 'planet_radius', 'stellar_effective_temperature'
, 'stellar_radius', 'transit_depth', 'transit_duration']
Yvals= 'status'


X= combined[Xvals]
Y= combined[Yvals]

X=np.array(X)
Y=np.array(Y)

#can't use strings to train the model, turn status into numbers
for i in range (len(Y)):
  if Y[i] in ['CANDIDATE', 'ACANDIDATE', 'APC']:
    Y[i]=0
  elif Y[i]=='CONFIRMED':
    Y[i]=1
  elif Y[i] in ['FALSE POSITIVE', 'FA']:
    Y[i]=2

print(X, Y)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd

# Clean Y and encode it
Y = pd.Series(Y).fillna("Unknown")         # replace NaN if any
le = LabelEncoder()
Y_enc = le.fit_transform(Y.astype(str))    # convert to string, encode

# Convert X, Y to numpy arrays
X = np.array(X)
Y_enc = np.array(Y_enc)

# Stratified split
Split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=69)
for train_index, test_index in Split.split(X, Y_enc):
    X_train = X[train_index]
    X_test = X[test_index]
    Y_train = Y_enc[train_index]
    Y_test = Y_enc[test_index]

# Train Random Forest
model = RandomForestClassifier(random_state=69)
model.fit(X_train, Y_train)
Yprediction = model.predict(X_test)

print("Predictions:", Yprediction)
print("Actual:", Y_test)
print("Accuracy:", 100 * accuracy_score(Y_test, Yprediction), "%")
```

```python
#Save the model
import joblib
joblib.dump(model, 'ExoPlanet_Classifier.joblib')
model=joblib.load('ExoPlanet_Classifier.joblib')
```

```python
# ['orbital_period', 'planet_radius', 'stellar_effective_temperature',
#  'stellar_radius', 'transit_depth', 'transit_duration']

print("Enter your ExoPlanetary Data:")

orbital_period = input("Orbital Period (Earth Days): ")
planet_radius = input("Planet Radius (Earth Radii): ")
stellar_effective_temperature = input("Stellar Effective Temperature (K): ")
stellar_radius = input("Stellar Radius (Sun Radii): ")
transit_depth = input("Transit Depth (ppm): ")
transit_duration = input("Transit Duration (Earth Hours): ")

ansnum = model.predict([[orbital_period, planet_radius, stellar_effective_temperature,
                         stellar_radius, transit_depth, transit_duration]])

breakdown = model.predict_proba([[orbital_period, planet_radius, stellar_effective_temperature,
                                  stellar_radius, transit_depth, transit_duration]])

if ansnum == 0:
    ansval = "candidate planet"
elif ansnum == 1:
    ansval = "confirmed planet"
elif ansnum == 2:
    ansval = "false positive"

confidence = float(breakdown[0][ansnum].item()) * 100
confidence = round(confidence, 2)
print("Your ExoPlanet's classification is a", ansval)
print("I am", confidence, "% sure in my prediction")
print("Breakdown: \n", "Candidate: " , str(100*breakdown[0][0]) , "%" ,
      "\n Confirmed: " , str(100*breakdown[0][1]) + "%", "\n False Positive: " , str(100*breakdown[0][2]) , "%")
```
