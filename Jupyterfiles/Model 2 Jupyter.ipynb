```python
# Check common columns (removed "ra" and "dec")
df1 = df1.drop(columns=["ra", "dec"], errors="ignore")
df2 = df2.drop(columns=["ra", "dec"], errors="ignore")

combined=pd.concat([df1, df2])

common_cols = set(df1.columns) & set(df2.columns)
print("Common columns after dropping ra and dec:")
print(sorted(common_cols))
```

```python
#Convert Data into a managable format for SciKit-Learn

Xvals= ['orbital_period', 'planet_radius', 'stellar_effective_temperature'
, 'stellar_radius', 'transit_depth', 'transit_duration']
Yvals= 'status'


X= combined[Xvals]
Y= combined[Yvals]

X=np.array(X)
Y=np.array(Y)

#can't use strings to train the model, turn status into numbers
for i in range (len(Y)):
  if Y[i] in ['CANDIDATE', 'ACANDIDATE', 'APC']:
    Y[i]=0
  elif Y[i]=='CONFIRMED':
    Y[i]=1
  elif Y[i] in ['FALSE POSITIVE', 'FA']:
    Y[i]=2

print(X, Y)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from xgboost import XGBClassifier


# Use a stratified shuffle split
Split= StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=69)
for train_index, test_index in Split.split(X, Y):
  X_train= X[train_index]
  X_test= X[test_index]
  Y_train= Y[train_index]
  Y_test= Y[test_index]


Y_train = Y_train.astype(int)
Y_test = Y_test.astype(int)

model = XGBClassifier(
    n_estimators=123,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    gamma=0.3,
    reg_lambda=1.0,
    num_class=100,
    objective='multi:softmax',
    eval_metric='mlogloss',
    use_label_encoder=False,
    random_state=69420
)


model.fit(X_train, Y_train)
Yprediction = model.predict(X_test)
print(Yprediction, Y_test)
print(100* accuracy_score(Y_test, Yprediction), "%")
```

```python
#Save the model
import joblib
joblib.dump(model, 'ExoPlanet_Classifier.joblib')
model=joblib.load('ExoPlanet_Classifier.joblib')
```

```python
7# ['orbital_period', 'planet_radius', 'stellar_effective_temperature',
#  'stellar_radius', 'transit_depth', 'transit_duration']

print("Enter your ExoPlanetary Data:")

orbital_period = input("Orbital Period (Earth Days): ")
planet_radius = input("Planet Radius (Earth Radii): ")
stellar_effective_temperature = input("Stellar Effective Temperature (K): ")
stellar_radius = input("Stellar Radius (Sun Radii): ")
transit_depth = input("Transit Depth (ppm): ")
transit_duration = input("Transit Duration (Earth Hours): ")

user_data = np.array([[orbital_period, planet_radius, stellar_effective_temperature,
                       stellar_radius, transit_depth, transit_duration]], dtype=np.float32)

ansnum = int(model.predict(user_data)[0])

breakdown = model.predict_proba(user_data)[0]

if ansnum == 0:
    ansval = "candidate planet"
elif ansnum == 1:
    ansval = "confirmed planet"
elif ansnum == 2:
    ansval = "false positive"

confidence = round(breakdown[ansnum] * 100, 2)
print("Your ExoPlanet's classification is a", ansval)
print("I am", confidence, "% sure in my prediction")
print("Prediction Breakdown: \n", "Candidate: " , str(100*breakdown[0]) , "%" ,
      "\n Confirmed: " , str(100*breakdown[1]) + "%", "\n False Positive: " , str(100*breakdown[2]) , "%")
```
